```python
import pandas as pd
import folium
import os
import geopandas
```

```python
spanishPopulationByProvince = os.path.join('data', 'dataw.csv')#9687sc.csv
spanishPopulation = pd.read_csv(spanishPopulationByProvince,converters={'ID': lambda x: str(x)})
```

Spanish geojson from https://github.com/codeforamerica/click_that_hood/tree/master/public/data

```python
provincesgeo = os.path.join('data', 'spain-provinces.geojson')
gdf = geopandas.read_file(provincesgeo)
```

```python
gdf
```

```python
m = folium.Map([40.4168, 3.7038], zoom_start=5)

folium.GeoJson(
    gdf,
).add_to(m)

m
```

```python
help(folium)
```

```python
m = folium.Map(location=[40, 3.5], zoom_start=5)
m.choropleth(
    geo_data=gdf,
    data=spanishPopulation,
    columns=['ID', 'Population'],
    key_on='feature.properties.cod_prov',
    fill_color='YlOrRd',#'YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.3,
    legend_name='Population',
    threshold_scale = [5e3,5e4,1e5,5e5,1e6,5e6],
    highlight=True
)

#m.save(os.path.join('results', 'Spain.html'))

m
```
